# **Funzioni di trasferimento**

```python
#################  RC  #################

def TrasfRC_VC(w, C, R):

  '''
  Modulo funzione di trasferimento per V_C in circuito RC
  '''

  #w: frequenza
  #C: capacità
  #R: resistenza (nota a priori)
  return 1 / np.sqrt(1 + (R*w*C)**2)

def Fase_TrasfRC_VC(w, C, R):

  '''
  Fase funzione di trasferimento per V_C in circuito RC
  '''

  #w: frequenza
  #C: capacità
  #R: resistenza (nota a priori)
  return - np.arctan(R*w*C)

def TrasfRC_VR(w, C, R):

  '''
  Modulo funzione di trasferimento per V_R in circuito RC
  '''

  #w: frequenza
  #C: capacità
  #R: resistenza (nota a priori)
  return (R*w*C) / np.sqrt(1 + (R*w*C)**2)

def Fase_TrasfRC_VR(w, C, R):

  '''
  Fase funzione di trasferimento per V_R in circuito RC
  '''

  #w: frequenza
  #C: capacità
  #R: resistenza (nota a priori)
  return np.pi/2 - np.arctan(R*w*C)

#################  RL  #################

def TrasfRL_VL(w, L, R):

  '''
  Modulo funzione di trasferimento per V_L in circuito RL
  '''

  #w: frequenza
  #L: induttanza
  #R: resistenza (nota a priori)
  return (w*L) / np.sqrt(R**2 + (w*L)**2)

def Fase_TrasfRL_VL(w, L, R):

  '''
  Fase funzione di trasferimento per V_L in circuito RL
  '''

  #w: frequenza
  #L: induttanza
  #R: resistenza (nota a priori)
  return np.pi/2 - np.arctan((w*L)/R)

def TrasfRL_VR(w, L, R):

  '''
  Modulo funzione di trasferimento per V_R in circuito RL
  '''

  #w: frequenza
  #L: induttanza
  #R: resistenza (nota a priori)
  return R / np.sqrt(R**2 + (w*L)**2)

def Fase_TrasfRL_VR(w, L, R):

  '''
  Fase funzione di trasferimento per V_R in circuito RL
  '''

  #w: frequenza
  #L: induttanza
  #R: resistenza (nota a priori)
  return - np.arctan((w*L)/R)

#################  RLC  #################

def TrasfRLC_VR(w, L, C, R):

  '''
  Modulo funzione di trasferimento per V_R in circuito RLC
  '''

  #w: frequenza
  #L: induttanza
  #C: Capacità
  #R: resistenza (nota a priori)

  return R / np.sqrt(R**2 + (w*L - 1/(w*C))**2)

def Fase_TrasfRLC_VR(w, L, C, R):

  '''
  Fase funzione di trasferimento per V_R in circuito RLC
  '''

  #w: frequenza
  #L: induttanza
  #C: Capacità
  #R: resistenza (nota a priori)

  return -np.arctan((w*L - 1/(w*C))/R)

def TrasfRLC_VL(w, L, C, R):

  '''
  Modulo funzione di trasferimento per V_L in circuito RLC
  '''

  #w: frequenza
  #L: induttanza
  #C: Capacità
  #R: resistenza (nota a priori)

  return (w*L)  / np.sqrt(R**2 + (w*L - 1/(w*C))**2)

def Fase_TrasfRLC_VL(w, L, C, R):

  '''
  Fase funzione di trasferimento per V_L in circuito RLC
  '''

  #w: frequenza
  #L: induttanza
  #C: Capacità
  #R: resistenza (nota a priori)

  return np.pi - np.arctan((w*L - 1/(w*C))/R)

def TrasfRLC_VC(w, L, C, R):

  '''
  Modulo funzione di trasferimento per V_C in circuito RLC
  '''

  #w: frequenza
  #L: induttanza
  #C: Capacità
  #R: resistenza (nota a priori)

  return (1 / (w*C)) / np.sqrt(R**2 + (w*L - 1/(w*C))**2)

def Fase_TrasfRLC_VC(w, L, C, R):

  '''
  Fase funzione di trasferimento per V_C in circuito RLC
  '''

  #w: frequenza
  #L: induttanza
  #C: Capacità
  #R: resistenza (nota a priori)

  return - np.pi / 2 - np.arctan((w*L - 1/(w*C))/R)
```

```python
def chi2(inter):
  return inter.fval

def ndof(misure, inter):
  return len(misure) - len(inter.values)

def p_value(misure, inter):
  return 1 - stats.chi2.cdf(chi2(inter), ndof(misure, inter))

def t_test(x_teorico, x_misurato, sigma_x_teorico, sigma_x_misurato):
  t = abs(x_misurato - x_teorico) / np.sqrt(sigma_x_teorico**2 + sigma_x_misurato**2)
  return t

def t_test2(t, ndof):
  return 1 - stats.t.cdf(t, ndof)
```
